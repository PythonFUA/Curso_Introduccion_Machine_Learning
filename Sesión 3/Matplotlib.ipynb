# **¿Qué es Matplotlib?**



matplotlib es la librería principal para la realización de gráficas científicas en Python. Esta librería proporciona funciones para la realizacion de gráficas como mapas, histogramas, gráficas de dispersión, entre otras. La visualización de sus datos y de diferentes aspectos de su analisis puede brindarle una mejor comprensión del problema a tratar. Para mayor información sobre la librería `matplotlib` puede consultar el tutorial de [Matplotlib](https://matplotlib.org/index.html).




1.   `matplotlib`: Cuando se habla de paquete, se habla de todo el paquete de visualización de Python.
2.   `pyplot` : corresponde a un módulo de `matplotlib`, el cual permite crear  gráficas de forma rápida y sin necesidad de instanciar figuras o ejes.

```
import matplotlib.pyplot as plt
import numpy as np
```

```
# Generar una secuencia de 100 numeros entre -10 y 10 
x = np.linspace(-10, 10, 100)
# Creamos un arreglo con el seno de éstos números
y = np.sin(x)
# Generamos una gráfica con x vs y
plt.plot(x, y, marker="x")
plt.title('Mi primera figura');
```

```
# Coordenadas x y de las gráficas para las funciones seno y coseno
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Gráfica de los puntos usando matplolib.
plt.plot(x, y_sin)
plt.plot(x, y_cos, color='k')
plt.xlabel('label del eje x')
plt.ylabel('label del eje y')
plt.title('Funciones seno y coseno')
plt.legend(['Seno', 'Coseno'])
plt.draw
plt.show()
```

## **Gráficas de variables categóricas.**

```
import matplotlib.pyplot as plt

names = ['groupo_a', 'groupo_b', 'groupo_c']
values = [5, 25, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Gráficas categóricas')
plt.show()
```

```
# Generación de los puntos (x,y) para la realización de los gráficos de 
# dispersión.
x=range(10)
y=np.random.rand(10)

plt.scatter(x,y)
plt.xlabel('label del eje x')
plt.ylabel('label del eje y')
plt.title('Gráfico de dispersión')
plt.show()
```

```
#personalice la gráfica

X = np.linspace(0,2*np.pi,5000) #Contruye un arreglo de 75 datos desde 0 hasta 2pi sin incluir el último
Y = np.sin(X)
fig, ax = plt.subplots()
ax.plot(X,Y,'--r')
ax.set_title('Función Seno') #Asignar un título a la gráfica
ax.set_xlabel('Tiempo, $t$') #Asignar un nombre al eje de las abscisas
ax.grid()
fig.show()
```

```
fig, ax = plt.subplots()
d=np.array([0.38, 0.64, 0.91, 1.26, 1.41, 1.66, 1.90, 2.18]) #arreglo de datos de la variable independiente
f=np.array([1.4, 1.65, 3.0, 3.95, 4.3, 5.20, 6.85, 7.4]) #arreglo de datos de la variable dependiente
df=np.array([ 0.4, 0.5, 0.4, 0.5, 0.6, 0.5, 0.5, 0.4]) #arreglo de datos de los errores en la variable dependiente
ax.errorbar(d,f,yerr=df,fmt='s')
ax.set_title('Fuerza vs distancia') #asigne un título adecuado a la gráfica
ax.set_xlabel('distancia (cm)') #asigne un nombre al eje de las abscisas
ax.set_ylabel('Fuerza (N)') #asigne un nombre al eje de las ordenadas
ax.grid()
ax.annotate("$x$",(1.25,5))
fig.show()
```
